Question 1: Implementing Linear Regression <br>

Description:<br>
Implement a simple linear regression model using Python and scikit-learn to predict house prices based on one feature: the size of the house.

```python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate a sample dataset
np.random.seed(0)
house_sizes = np.random.rand(100, 1) * 1000  # in square feet
house_prices = 200 + 0.5 * house_sizes + np.random.randn(100, 1) * 10000  # in dollars

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(house_sizes, house_prices)

# Get the coefficients
intercept = model.intercept_[0]
coefficient = model.coef_[0][0]

print(f"House Price = ${intercept:.2f} + ${coefficient:.2f} * House Size (sq ft)")

# Make predictions
new_house_size = [[1500]]  # in square feet
predicted_price = model.predict(new_house_size)
print(f"Predicted price for a {new_house_size[0][0]} sq ft house: ${predicted_price[0][0]:.2f}")

# Plot the data and the regression line
plt.scatter(house_sizes, house_prices, label='Data')
plt.plot(house_sizes, model.predict(house_sizes), color='red', label='Regression Line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($)')
plt.title('House Price vs House Size')
plt.legend()
plt.show()
```
