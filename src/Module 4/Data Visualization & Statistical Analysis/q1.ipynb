Title: Histogram of a Single Feature <br>

Question 1: Create a histogram for the age feature from a dataset. Interpret what the shape of the histogram tells us about the distribution of the age feature.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a sample dataset
np.random.seed(0)
ages = np.random.randint(18, 80, size=100)
df = pd.DataFrame({'age': ages})

# Create a histogram for the age feature
plt.hist(df['age'], bins=10, edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

Title: Scatter Plot to Analyze Relationship<br>

Question 4: Create a scatter plot for the weight and height features to determine if there is a trend.

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a sample dataset
np.random.seed(0)
heights = np.random.normal(175, 8, size=100)  # Mean height in cm
weights = heights * 0.5 + np.random.normal(0, 5, size=100) + 50  # Weight in kg
df = pd.DataFrame({'height': heights, 'weight': weights})

# Create a scatter plot for the weight and height features
plt.figure(figsize=(8, 6))
plt.scatter(df['height'], df['weight'])
plt.title('Scatter Plot of Weight vs Height')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()

```

Title: Correlation Heatmap<br>

Question 5 : Generate a correlation heatmap for a dataset with multiple features (e.g., height ,weight , age ) and explain the correlations observed.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a sample dataset
np.random.seed(0)
heights = np.random.normal(175, 8, size=100)  # Mean height in cm
weights = heights * 0.5 + np.random.normal(0, 5, size=100) + 50  # Weight in kg
ages = np.random.randint(18, 80, size=100)
df = pd.DataFrame({'height': heights, 'weight': weights, 'age': ages})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Generate a correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()


```

Title: Descriptive Statistical Analysis<br>

Question 6: Calculate the mean, median, standard deviation, skewness, and kurtosis for the
temperature feature and discuss the results.

```python

import pandas as pd
import numpy as np

# Create a sample dataset
np.random.seed(0)
temperatures = np.random.normal(25, 5, size=100)  # Mean temperature in Celsius
df = pd.DataFrame({'temperature': temperatures})

# Calculate descriptive statistics
mean = df['temperature'].mean()
median = df['temperature'].median()
std_dev = df['temperature'].std()
skewness = df['temperature'].skew()
kurtosis = df['temperature'].kurtosis()

print("Descriptive Statistics:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

Title: Covariance and Correlation between Two Features<br>

Question 7: Compute the covariance and correlation between price and demand in a dataset.
Explain what these metrics indicate.

```python
import pandas as pd
import numpy as np

# Create a sample dataset
np.random.seed(0)
prices = np.random.uniform(10, 50, size=100)
demands = 100 - prices * 0.5 + np.random.normal(0, 5, size=100)
df = pd.DataFrame({'price': prices, 'demand': demands})

# Compute covariance
covariance = df['price'].cov(df['demand'])
print(f"Covariance: {covariance:.2f}")

# Compute correlation
correlation = df['price'].corr(df['demand'])
print(f"Correlation: {correlation:.2f}")
```

Title: Principal Component Analysis (PCA)<br>

Question 9 : Apply PCA on a dataset with multiple features (e.g., x1 , x2 , x3 , x4 ) and reduce it to two principal components. Visualize the data in the new feature space.

```python


import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Create a sample dataset
np.random.seed(0)
x1 = np.random.normal(0, 1, size=100)
x2 = np.random.normal(1, 2, size=100)
x3 = x1 * 0.5 + np.random.normal(0, 1, size=100)
x4 = x2 * 0.7 + np.random.normal(0, 1, size=100)
df = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4})

# Apply PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df)

# Create a new DataFrame with the principal components
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])

# Visualize the data in the new feature space
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.title('Data in the New Feature Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

```

Title: Advanced Pair Plot with Hue Parameter<br>

Question 10 : Create a pair plot for height , weight , and age with an added categorical variable gender as the hue to observe different group trends.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a sample dataset
np.random.seed(0)
heights_male = np.random.normal(180, 8, size=50)  # Mean height in cm for males
heights_female = np.random.normal(165, 7, size=50)  # Mean height in cm for females
weights_male = heights_male * 0.5 + np.random.normal(0, 5, size=50) + 60  # Weight in kg for males
weights_female = heights_female * 0.5 + np.random.normal(0, 5, size=50) + 50  # Weight in kg for females
ages = np.random.randint(18, 80, size=100)

# Create a DataFrame
heights = np.concatenate([heights_male, heights_female])
weights = np.concatenate([weights_male, weights_female])
genders = ['Male'] * 50 + ['Female'] * 50
df = pd.DataFrame({'height': heights, 'weight': weights, 'age': ages, 'gender': genders})

# Create a pair plot with hue
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='gender')
plt.show()
```
