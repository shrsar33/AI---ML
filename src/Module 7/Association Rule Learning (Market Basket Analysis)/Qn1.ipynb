```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Task 1: Grocery Store Transactions
transactions = [
    ['Milk', 'Processing', 'Bread'],
    ['Milk', 'Diaper', 'Beer', 'Bread'],
    ['Milk', 'Diaper', 'Beer', 'Coke'],
    ['Bread', 'Egg', 'Milk'],
    ['Bread', 'Egg', 'Diaper', 'Milk', 'Beer']
]

# Convert the transactional data into a pandas DataFrame
itemsets = []
for transaction in transactions:
    itemset = {}
    for item in set([item for sublist in transactions for item in sublist]):
        itemset[item] = 1 if item in transaction else 0
    itemsets.append(itemset)

df = pd.DataFrame(itemsets)

# Apply the Apriori algorithm
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

print("Frequent Itemsets:")
print(frequent_itemsets)

# Task 2: Retail Store Data
transactions = [
    ['Shirt', 'Tie'],
    ['Shirt', 'Belt', 'Tie'],
    ['Shirt', 'Belt'],
    ['Tie', 'Belt'],
    ['Shirt', 'Tie', 'Belt']
]

# Convert the transactional data into a pandas DataFrame
itemsets = []
for transaction in transactions:
    itemset = {}
    for item in set([item for sublist in transactions for item in sublist]):
        itemset[item] = 1 if item in transaction else 0
    itemsets.append(itemset)

df = pd.DataFrame(itemsets)

# Apply the Apriori algorithm
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

print("\nAssociation Rules:")
print(rules)

# Task 3: Bookstore Purchases
transactions = [
    ['Book A', 'Book B'],
    ['Book A', 'Book C'],
    ['Book B', 'Book C', 'Book A'],
    ['Book B', 'Book D']
]

# Convert the transactional data into a pandas DataFrame
itemsets = []
for transaction in transactions:
    itemset = {}
    for item in set([item for sublist in transactions for item in sublist]):
        itemset[item] = 1 if item in transaction else 0
    itemsets.append(itemset)

df = pd.DataFrame(itemsets)

# Apply the Apriori algorithm
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

print("\nAssociation Rules:")
print(rules)
```
